```python
import sys; sys.path.insert(0,'../src')
import numpy as np
from models.graphene import SingleLayerGrapheneTB
from response.conductivity import optical_conductivity,optical_conductivity_xx
```

```python
tb=SingleLayerGrapheneTB(t=2.78)
omega=np.linspace(0.0,3.0,150)+1e-6
domega = omega[1]-omega[0]
eta = 0.12 #domega*(1+1/np.sqrt(3))*13
res=optical_conductivity(tb,omega,nk=200,eta=eta,Ef=0)
sig_i=res['inter'][:,0,0].real
sig_t=res['total'][:,0,0].real
sig_d=res['intra'][:,0,0].real
m=(omega>0.75)&(omega<2.5)
print(f'<Re[sigma_inter]>={np.mean(sig_i[m]):.4f}')
print(f'<Re[sigma_total]>={np.mean(sig_t[m]):.4f}')
print(f'Re[sigma_intra](DC)={sig_d[0]:.4f} (Ef=0 -> ~0)')
print(f'Expected universal: 0.25 = e^2/(4 hbar)')
r=optical_conductivity_xx(tb,omega,nk=200,eta=eta,Ef=0)
print(f'xx wrapper <Re[sigma]>={np.mean(r["total"][(omega>0.75)&(omega<2.5)]):.4f}')
```

```python

import matplotlib.pyplot as plt

plt.style.use('bmh')

%matplotlib inline
%config InlineBackend.figure_format='retina'
import matplotlib as mpl
mpl.rcParams.update({
    'font.size': 10.5,
    # 'font.family': ['Times New Roman','SimSun'],# 'HP Simplified Hans', # Tahoma
    'axes.unicode_minus': False,
    'axes.facecolor': 'white',
    'figure.facecolor': 'white',
    'savefig.facecolor': 'white', # optional: saved figures
    # 'grid.color': '#e0e0e0', # optional: lighter grid
    'axes.edgecolor': 'black', # optional: clearer axes
})

fig, ax = plt.subplots(figsize=(4,3))
ax.plot(omega, sig_t)
# ax.plot(omega, r["total"], '--') # 3.3e7 for i=31; 8e6 for i=15
ax.set_xbound(omega[0], omega[-1])
ax.set_xlabel(r"Energy (eV)") # $\gamma_0$
ax.set_ylabel("Opt. Cond.") #  (states/eV·unit cell)
ax.set_title(''.join([r'$\eta$=', f'{eta:.2g} ',r'eV; $\Delta E$=',f'{domega:.3g} ','eV']), loc='right') #$\gamma_0$
fig.tight_layout()
fig.savefig('Graphene-Universal-Conductivity.png',dpi=200)
plt.show()
```
